```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

%matplotlib inline
```

```python
def draw(qc):
    qc.draw(output='mpl')
```

```python
circuit = QuantumCircuit(3, 3)
draw(circuit)
```

```python
circuit.x(0)

circuit.barrier()
draw(circuit)
```

```python
circuit.h(1)

circuit.cx(1, 2)
draw(circuit)
```

```python
circuit.cx(0, 1)
circuit.h(0)

circuit.barrier()

circuit.measure([0, 1], [0, 1])

draw(circuit)
```

```python
circuit.barrier()

circuit.cx(1, 2)
circuit.cz(0, 2)

draw(circuit)
```

```python
from qiskit.visualization import plot_histogram

sim = AerSimulator(method='automatic')


circuit.measure(2, 2)

result = sim.run(circuit).result()
counts = result.get_counts()

plot_histogram(counts);
```
