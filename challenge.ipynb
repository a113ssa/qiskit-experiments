```python
from qiskit.visualization import plot_bloch_multivector, array_to_latex
from custom_utils import create_circuit, run_simulation

%matplotlib inline
```

1101 binary number representation by gates application to qubits:

```python
circuit = create_circuit(4)
circuit.draw(output='mpl');
```

```python
INDEXES_OF_1_BIT = [0, 2, 3]
```

```python

# with X gates
for index in INDEXES_OF_1_BIT:
    circuit.x(index)

result = run_simulation(circuit)
output_state_x = result.get_statevector()
plot_bloch_multivector(output_state_x);
```

```python
circuit = create_circuit(4)

# with Y gates
for index in INDEXES_OF_1_BIT:
    circuit.y(index)

result = run_simulation(circuit)
output_state_y = result.get_statevector()
plot_bloch_multivector(output_state_y);
```
