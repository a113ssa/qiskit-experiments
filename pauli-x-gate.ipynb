```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector
from qiskit_aer import AerSimulator
from math import pi

%matplotlib inline
```

```python
def create_circuit(qubits=1):
    circuit = QuantumCircuit(qubits)
    circuit.ry(pi/4, 0)
    return circuit

def run_simulation(circ):
    simulator = AerSimulator(method='statevector')
    circ.save_statevector()
    result = simulator.run(circ).result()
    return result
```

```python
# create a quantum circuit with one qubit
circuit_before_x = create_circuit()
circuit_before_x.draw(output='mpl');
```

```python
# before X Gate
result_1 = run_simulation(circuit_before_x)
output_state_1 = result_1.get_statevector()
plot_bloch_multivector(output_state_1);
```

```python
# after X Gate
circuit_after_x = create_circuit()
circuit_after_x.x(0)
result_2 = run_simulation(circuit_after_x)
output_state_2 = result_2.get_statevector()
plot_bloch_multivector(output_state_2);
```
